# Limpieza de datos

Después de realizar el EDA veo que hay columnas con muchos valores nulos.

```python
#Visualizo la cantidad de nulos por columna con un gráfico de barras
import matplotlib.pyplot as plt

#Cuento los valores nulos en cada columna
nulos = bac_anual_df.isnull().sum()

# Filtro columnas que tienen valores nulos
nulos = nulos[nulos > 0]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
nulos.plot(kind='bar')
plt.title('Número de valores nulos por columna')
plt.xlabel('Columnas')
plt.ylabel('Número de valores nulos')
plt.xticks(rotation=90)
plt.show()
```

```python
#nulos.count()
```

```python
#Elimino aquellas columnas que tienen mas del 50% de sus valores nulos
thresh = bac_anual_df.shape[0] * 0.5
bac_limpio = bac_anual_df.dropna(axis=1, thresh=thresh)
bac_limpio.shape
```

```python
bac_limpio.isnull().sum()
```

Observo que siguen quedando columnas con una consdireable cantidad de valores nulos y que parecen no ser relevantes para el análisis. Como por ejemplo el mail del punto de contacto

```python
#Pruebo quedarme con las columnas con mas del 70% de sus valores no nulos
thresh = bac_anual_df.shape[0] * 0.7
bac_limpio = bac_anual_df.dropna(axis=1, thresh=thresh)
bac_limpio.shape
```

```python
#Veo que bajo considerablemente la cantidad de columnas y que siguen quedan algunos valores nulos
bac_limpio.isnull().sum()
```

```python
#Aplico un script para eliminar filas con el 50% de sus calores nulos y veo que no hub cambios
thresh_value = bac_limpio.shape[1] - 50
bac_limpio = bac_limpio.dropna(axis=0, thresh=thresh_value)
bac_limpio.shape
```

```python
# Observo la distribución de valores en la columna con mas valores nulos
value_counts = bac_limpio['tender/status'].value_counts(dropna=False)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
value_counts.plot(kind='bar')
plt.title('Distribución de Estados de Licitación')
plt.xlabel('Estado')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.show()
```

```python
#Remplazo los valores nulos de las columnas categóricas por la moda
#moda= bac_limpio['parties/0/name'].mode()[0]

for col in bac_limpio.select_dtypes(include=['object']):
    bac_limpio[col].fillna(bac_limpio[col].mode()[0], inplace=True)
```

```python
#Ya no tengo valores nulos
bac_limpio.isnull().sum()
```

```python
#Convierto las columnas de fechas a objetos de tipo datetime y elimino timezone
def a_fecha(dataframe, columna):
  dataframe[columna] = pd.to_datetime(dataframe[columna], errors = 'coerce') #pasando a datetime
  dataframe[columna] = dataframe[columna].dt.tz_localize(None)  #eliminando timezone

a_fecha(bac_limpio,'date')
a_fecha(bac_limpio,'tender/tenderPeriod/startDate')
a_fecha(bac_limpio,'tender/tenderPeriod/endDate')
a_fecha(bac_limpio,'tender/enquiryPeriod/startDate')
a_fecha(bac_limpio,'tender/enquiryPeriod/endDate')
a_fecha(bac_limpio,'tender/documents/0/datePublished')
```

```python
#Creo otro DataFrame sin las columnas Id, lo que puede ser útil para modelos de aprendizaje automático
columnas_id = ['ocid','id', 'tender/id','tender/procuringEntity/id','tender/items/0/id','tender/items/0/classification/id','tender/documents/0/id','parties/0/id','parties/0/identifier/id']
bac_limpio2 = bac_limpio.drop(columns=columnas_id)
bac_limpio2.head(3)
```

```python
bac_limpio2.dtypes
```

```python
#Normalizo las 3 columnas numéricas del dataframe y guardo los datos antes de normalizar para una visualización
from sklearn.preprocessing import StandardScaler
columnas_numericas = ['tender/value/amount', 'tender/items/0/quantity', 'tender/items/0/unit/value/amount']
# Copiar los datos originales antes de la normalización
datos_antes = bac_limpio2[columnas_numericas].copy()

# Normalizar las columnas numéricas
scaler = StandardScaler()
bac_limpio2[columnas_numericas] = scaler.fit_transform(bac_limpio2[columnas_numericas])

# Copiar los datos después de la normalización
datos_despues = bac_limpio2[columnas_numericas].copy()
```

```python
# Creo una visualización comparativa de las densidades antes y despues de normalizar
import seaborn as sns
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(15, 10))

# Gráficos de densidad antes de la normalización
for columna in columnas_numericas:
    sns.kdeplot(data=datos_antes, x=columna, ax=axes[0], label=columna)
axes[0].set_title('Densidad antes de la normalización')
axes[0].legend()

# Gráficos de densidad después de la normalización
for columna in columnas_numericas:
    sns.kdeplot(data=datos_despues, x=columna, ax=axes[1], label=columna)
axes[1].set_title('Densidad después de la normalización')
axes[1].legend()

plt.tight_layout()
plt.show()
```
